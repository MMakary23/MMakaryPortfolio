## The Data (15 marks)


The dataset used in the project comes from the Dubai Land Department (DLD), which was founded in 1960 with the aim of establishing the most prominent real estate sector at regional and international levels. The Dubai Land Department (DLD) is a Dubai Government entity responsible for the registration, organization, and promotion of real-estate investments in Dubai, and therefore has data on all the real estate transactions done in the city over the years.

The dataset contains all the real estate transactions made in Dubai since 1975, this includes sales, mortgage, inheritance, merge, separation and sub-separation. The property types included in the dataset are villas, lands, buildings and units. 1.1 million transactions and 46 attributes make up the dataset. Some of the columns were available in English and Arabic, but the Arabic columns were dropped.

In addition, a number of columns were dropped due to their irrelevancy to the objectives of the project. The 19 attributes that are used for the analysis after dropping the irrelevant ones are transaction_id, instance_date, property_type_id, property_type, property_sub_type, property_usage, reg_type_id, reg_type, area_id, area_name, building_name, nearest_landmark, nearest_metro, nearest_mall, rooms, has_parking, procedure_area, actual_worth, meter_sale_price. A number of the rows were also dropped since only transactions made from the year 2000 onwards were included; transactions concerning lands, buildings and villas were excluded, and only the ones related to units remained since the scope of analysis is only apartments, the same thing applies for the rooms column where only Studios, 1, 2 and 3 bedrooms were included excluding the rest; and finally, only properties usage of residential were included in the data frame used for analysis.

In terms of data quality, no duplicates were found during the exploratory data analysis. However, a large number of missing values were found under the two columns related to the rent value of different units, but these columns were completely dropped since the analysis objectives don't involve the rent value. The column 'property_sub_type' has 250,000 missing values but EDA showed that they are all belonging to the 'Villa' type since it doesn't have any subtypes, the subtypes column was included for the 'Unit type properties. The 3 columns nearest_landmark, nearest_mall and nearest_metro all have missing values, since not all properties have a landmark, mall, or metro in near proximity, and these missing values actually contribute to the accuracy of the data and are useful for the analysis, and based on that they are left as is.

During the analysis, it became evident that certain columns, namely procedure_area and meter_sale_price, contained erroneous values. To address this, logical and realistic minimum and maximum thresholds were established for these columns. Similarly, the "actual_worth" column was refined to include only properties valued between 500k and 5 million. This decision was taken considering Dubai's broad range of property prices and to ensure that our analysis and recommendations were grounded in a realistic and relevant price bracket for most of the users. Any values outside these determined bounds in any of the columns were treated as outliers and subsequently omitted. These thresholds and value brackets were informed by real-time data from current Dubai apartment listings and a comprehensive understanding of the city's real estate landscape.

#### The Boxplots clearly illustrates that there are outliers and wrong values. The outliers will be dealth with by setting thresholds and value brackets informed by real-time data from current Dubai apartment listings and a comprehensive understanding of the city's real estate landscape.

```python
# Set Thresholds of meter squared area of differnt kinds of apartments
area_limits = {
    'Studio': (30, 125),
    '1 B/R': (60, 230),
    '2 B/R': (70, 360),
    '3 B/R': (100, 500)
}

# List to collect outlier indices
outliers_indices = []

# Flag or Filter Outliers
for room_type, (min_area, max_area) in area_limits.items():
    outliers_area = df_apartments[(df_apartments['rooms'] == room_type) & 
                                  ((df_apartments['procedure_area'] <= min_area) | 
                                   (df_apartments['procedure_area'] >= max_area))]
    print(f"Area Outliers for {room_type}: {len(outliers_area)} entries")
    
    # Collect indices of the outliers
    outliers_indices.extend(outliers_area.index.tolist())

# Drop the outliers
df_apartments.drop(outliers_indices, inplace=True, errors='ignore')
```

```python
# Set Price Bounds of meter price of different kinds of apartments
price_bounds = {
    'Studio': (8300, 55000),
    '1 B/R': (5750, 56000),
    '2 B/R': (5000, 90000),
    '3 B/R': (5500, 33000),
}

# List to collect outlier indices for meter price
outliers_meter_price_indices = []

# Flag or Filter Outliers
for room, (min_price, max_price) in price_bounds.items():
    outliers_meter_price = df_apartments[(df_apartments['rooms'] == room) & 
                                         ((df_apartments['meter_sale_price'] < min_price) | 
                                          (df_apartments['meter_sale_price'] > max_price))]
    print(f"Meter Price Outliers for {room}: {len(outliers_meter_price)} entries")
    
    # Collect indices of the meter price outliers
    outliers_meter_price_indices.extend(outliers_meter_price.index.tolist())

# Drop the meter price outliers
df_apartments.drop(outliers_meter_price_indices, inplace=True, errors='ignore')
```

#### Recommendation functions that help apartments buyers in Dubai decide on which areas to include in their search by showing the top 25 recommended areas

```python
def budget_condition(budget):
    return df_recommender['actual_worth'] <= int(budget)
```

```python
def room_condition(number_of_rooms):
    return df_recommender['rooms'] == number_of_rooms
```

```python
def metro_condition(near_metro):
    if near_metro == 'yes':
        return df_recommender['nearest_metro'] != 'no metro'
    else:
        return df_recommender['nearest_metro'] == 'no metro'
```

```python
def parking_condition(has_parking):
    if has_parking == 'yes':
        return df_recommender['has_parking'] == True
    else:
        return df_recommender['has_parking'] == False
```

```python
table_style = """
<style>
    table.dataframe {
        border-collapse: collapse;
        border: 1px solid black;
    }
    table.dataframe th {
        border: 1px solid black;
        background-color: skyblue;
        padding: 8px;
        text-align: left;
    }
    table.dataframe td {
        border: 1px solid black;
        padding: 8px;
    }
</style>
"""
```

```python
last_recommendations = []

def recommend_apartments():
    # Get user input
    budget = float(input("Enter your budget: "))
    number_of_rooms = input("Enter the number of rooms desired (Studio, 1 B/R, 2 B/R, or 3 B/R): ")
    near_metro = input("Do you prefer being near a metro station? (yes/no): ").lower()
    has_parking = input("Do you need parking? (yes/no): ").lower()
    
    # Apply the conditions
    condition = (budget_condition(budget) & 
                 room_condition(number_of_rooms) & 
                 metro_condition(near_metro) & 
                 parking_condition(has_parking))
    
    global last_recommendations  # Declare the variable as global to modify it
    filtered_df = df_recommender.loc[condition]
    last_recommendations = filtered_df['area_name'].tolist()
    
    # Get top 25 areas with their counts
    area_counts = pd.Series(last_recommendations).value_counts().head(25)
    df = area_counts.reset_index()
    df.columns = ["Recommended Areas", "Count"]
    
    # Display the areas and their counts as an HTML table
    display(HTML(table_style + df.to_html(index=False)))
```

#### Recommendation functions that help apartments sellers decide on a price for their soon-to-be-listed apartment.

```python
# Definitions for conditions
def room_condition(room):
    return df_apartments['rooms'] == room
```

```python
def area_name_condition(area_name):
    return df_apartments['area_name'] == area_name
```

```python
def procedure_area_condition(procedure_area):
    return df_apartments['procedure_area'].between(procedure_area - 5, procedure_area + 5)  # example range
```

```python
def has_parking_condition(has_parking):
    return df_apartments['has_parking'] == has_parking
```

```python
def nearest_metro_condition(nearest_metro):
    if nearest_metro == 'yes':
        return df_apartments['nearest_metro'] != 'no metro'
    else:
        return df_apartments['nearest_metro'] == 'no metro'
```

```python
# Input Functions
def input_rooms():
    return input("Enter the number of rooms desired (Studio, 1 B/R, 2 B/R, or 3 B/R): ")
```

```python
def input_area_name():
    area_dropdown = widgets.Dropdown(
        options=sorted(df_apartments['area_name'].unique().tolist()),
        description='Area:',
        disabled=False,
    )
    display(area_dropdown)
    return area_dropdown.value
```

```python
def input_procedure_area():
    return float(input("Enter the procedure area in square meters: "))
```

```python
def input_has_parking():
    return int(input("Is there a parking provided with the apartment? (yes/no): ").lower() == "yes")
```

```python
def input_nearest_metro():
    return input("Is there a metro station nearby? (yes/no): ").lower()
```

```python
def filter_data(rooms, area_name, procedure_area, has_parking, nearest_metro):
    condition = (room_condition(rooms) & 
                 area_name_condition(area_name) & 
                 procedure_area_condition(procedure_area) & 
                 has_parking_condition(has_parking) & 
                 nearest_metro_condition(nearest_metro))
    
    return df_apartments[condition]
```

```python
def calculate_prices(df):
    return df['actual_worth'].min(), df['actual_worth'].mean(), df['actual_worth'].max()
```

```python
def display_prices(min_price, avg_price, max_price):
    result = f"""
    <div style='background-color: #f5f5f5; padding: 10px; border-radius: 5px; width: 50%;'>
        <h3>Recommended Selling Price:</h3>
        <p><b>Min Price:</b> {min_price}</p>
        <p><b>Average Price:</b> {avg_price}</p>
        <p><b>Max Price:</b> {max_price}</p>
    </div>
    """
    display(HTML(result))
```

```python
def recommend_selling_price():
    rooms = input_rooms()
    area_name = input_area_name()
    procedure_area = input_procedure_area()
    has_parking = input_has_parking()
    nearest_metro = input_nearest_metro()

    filtered_data = filter_data(rooms, area_name, procedure_area, has_parking, nearest_metro)

    if not filtered_data.empty:
        min_price, avg_price, max_price = calculate_prices(filtered_data)
        display_prices(round(min_price, 2), round(avg_price, 2), round(max_price, 2))
    else:
        print("No data available based on the given inputs.")
```

```python
recommend_selling_price()
```
